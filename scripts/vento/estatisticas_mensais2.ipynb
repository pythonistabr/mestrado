```python
import locale
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('default')
```

```python
locale.setlocale(locale.LC_ALL, 'en_US.utf8')
```

```python
plt.rcParams['font.weight'] = 'bold'
plt.rcParams['font.size'] = 10
plt.rcParams['lines.linewidth'] = 2
plt.rcParams["legend.title_fontsize"] = 20
plt.rcParams["lines.markersize"] = 10
```

```python
path = '/home/gus/Documentos/mestrado/dados-mestrado/vento-inmet/'
```

```python
dfwind = pd.read_csv(path+'vento_medias_mensais.csv', delimiter=';')
dfwind_std = pd.read_csv(path+'vento_std_mensal.csv', delimiter=';')
```

```python
dfwind.head(3)
```

```python
dfwind.direction = (dfwind.direction - 180) % 360
```

```python
dfwind
```

```python
months = ['Dec-16', 'Jan-17', 'Feb-17', 'Mar-17-','Apr-17', 'May-17',
          'Jun-17','Jul-17','Aug-17','Sep-17','Oct-17','Nov-17','Dez-17']
```

```python

fig, ax1 = plt.subplots(figsize=(8,5), dpi=300)

ax1.errorbar(months, dfwind.direction, dfwind_std.direction/2,
             fmt='--o',c='#6497b1',capsize=10)

ax1.set_ylabel('DIRECTION($\circ$)')

ax1.set_xticklabels(months,rotation = 15);

ax1.set_xlabel('\n TIME(MONTHS)')

#plt.legend()
```

```python
fig, ax1 = plt.subplots(1,1, figsize=(8,5),dpi=300)

ax1.errorbar(months, dfwind.w, dfwind_std.w/2,
             fmt='--s',c='orangered',capsize=10,label='Wind magnitude')

ax1.errorbar(months, dfwind.wg, dfwind_std.wg/2,
             fmt='--s',c='#6497b1',capsize=10, label='Wind gust')


ax1.set_ylabel('VELOCITY ($ m \cdot s^{-1}$)')

ax1.set_xticklabels(months,rotation = 15);

ax1.set_xlabel('\n TIME(MONTHS)')

plt.legend()
```
