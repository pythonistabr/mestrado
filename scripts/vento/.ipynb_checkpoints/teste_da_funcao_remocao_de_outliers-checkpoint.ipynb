```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from filters import remove_outliers
```

```python
%matplotlib inline
```

## criando um dataset com um dado a cada minuto

```python
tempo = pd.date_range(start='2022-01-01 00:00',\
                      end='2022-01-01 23:59', freq='1min')
```

```python
x = np.arange(-2*np.pi, 2*np.pi, (4*np.pi)/len(tempo)) # variavel x

y = np.sin(3*x) + np.cos(x/2) # variavel y
```

```python
# dataset
dados = pd.DataFrame({'x': x, 'y': y, 'tempo': tempo})
```

```python
# plot do dado sem outliers
plt.plot(dados.tempo, dados.y,'.')
```

### adicionando 50 outliers na variavel y do dataset  "dados"

```python
import random
```

```python
for ii in range(50):
    dados.iloc[random.randint(0,1440), 1] = random.randint(10, 20)
```

```python
plt.plot(dados.tempo, dados.y,'.')
```

### usando a funcao remove_outliers com uma janela de 15 minutos pra limpar o dataset

```python
"""
aqui eu estou filtrando os dados em um intervalo de 30 minutos
"""

yFiltrado = remove_outliers(dados.y, 15)
```

## plot dos dados "sujos"

```python
plt.plot(tempo, dados.y,'.b')
```

## plot dos dados limpos

```python
plt.plot(tempo, yFiltrado,'.r')
```
