<h2> Equação para encontrar as coponentes u e v do vento (na orientação oceanográfica).</h2>
<h3>$ \phi_{m}+180 = \phi_{o}$</h3>
<h3> $\phi_{u} = 90 - \phi_{o} \therefore \phi_{u} = -90 - \phi_{m}$</h3>
<h3> $Wv = r\cdot\sin(\phi_{u})$ </h3>
<h3> $Wu = r\cdot\cos(\phi_{u})$</h3>
<br>
<br>
<p>$\phi_{m}$ = direção meteorologica </p>
<p>$\phi_{o}$ = direção oceanografica </p>
<p>Wu =  componente zonal</p>
<p>Wv =  componente meridional</p>

<h3>Funcoes</h3>

```python
def calcula_uv(magnitude, direcao):
    
    """
    # função para calcular as componentes
    # u e v do vento na orientação oceano-
    # grafica.
    """
    from numpy import deg2rad, cos, sin
    from numpy import pi, array
    
    phi_u = 90 - direcao
    phi_u[phi_u<0] += 360
    phi_u = deg2rad(phi_u)
    
    u = magnitude * cos(phi_u)
    v = magnitude * sin(phi_u)
    values = {"u": u, "v":v}
    return values
```

```python
def meteo2ocean(angles):
    """
    converte um dado de vento dado na 
    convencao meteorologica para a
    oceanografica
    """
    angles = (angles + 180.0) % 360
    return angles
```

```python
def rotaciona_vento(angulo, valor):
    angulo += valor
    angulo[angulo < 0] += 360
    return angulo
```

```python
def comp2direction(x, y):
    
    phi_u = np.arctan2(y, x)
    
    phiOcean = 0.5*np.pi - phi_u
    
    phiOcean = np.rad2deg(phiOcean)
    
    phiOcean[phiOcean<0] += 360
   
    return phiOcean
```

### componentes paralela e perpendicular do vento

```python
values = calcula_uv(df.w, direction)
values_gust = calcula_uv(df.wg, direction)
```

```python
# -- componentes zonal e meridional --

wu, wv = values["u"], values["v"]
wgu, wgv = values_gust["u"], values_gust["v"]
```

```python
wul = wu*np.cos(np.deg2rad(60)) + wv*np.sin(np.deg2rad(60))
wvl = -wu*np.sin(np.deg2rad(60)) + wv*np.cos(np.deg2rad(60))


wgul = wgu*np.cos(np.deg2rad(60)) + wgv*np.sin(np.deg2rad(60))
wgvl = -wgu*np.sin(np.deg2rad(60)) + wgv*np.cos(np.deg2rad(60))
```

```python
df_rotated = pd.DataFrame({"wul": wul, "wvl": wvl,
             "wgul":wgul, "wgvl":wgvl})
```

### Calculo das estatisticas mensais

<p style='color:black; font-size:20px' >Valores mensais de vento
</p>

<p>Valores minimos, maximos, medios e desvio padrão das direções de vento
para todo o periodo</p>

```python
rc = df.copy()
```

```python
rc['direction'] = meteo2ocean(rc['direction'])
```

```python
values = calcula_uv(rc.w, rc.direction)
gust_values = calcula_uv(rc.wg, rc.direction)
```

```python
rc['wu'] = values['u']
rc['wv'] = values['v']
```

```python
rc['wgu'] = gust_values['u']
rc['wgv'] = gust_values['v']
```

```python
del values, gust_values
```

```python
rc2 = rc.groupby(['year','month']).mean().reset_index()
rc2_std = rc.groupby(['year','month']).std().reset_index()
```

```python
rc2.drop(['day','hour'], axis=1, inplace=True)
rc2_std.drop(['day','hour'], axis=1, inplace=True)
```

```python
rc2.drop('direction', axis=1, inplace=True)
```

```python
rc2_std.drop('direction', axis=1, inplace=True)
```

<h4>calculando as direções pelas componentes</h4>

```python
direction = comp2direction(rc2.wu, rc2.wv)
direction_std = comp2direction(rc2_std.wu, rc2_std.wv)

direction[direction<0] += 360.0
direction_std[direction_std<0] += 360.0
```

```python
rc2['direction'] = direction
rc2_std['direction'] = direction_std


#rc2['direction'] = rc2['direction'] - 180.0
#rc2['direction'].loc[rc2.direction<0]+= 360


del direction, direction_std
```
