```python
df = pd.read_csv('vento_filtrado.csv', delimiter=';')
```

```python
df.head(3)
```

```python
Time_in_datetime = [dt.strptime(ii, '%Y-%m-%d %H:%M:%S') for ii in df.time]
```

```python
Time_in_datetime = pd.Series(Time_in_datetime)
```

```python
df.drop(['time'],inplace=True, axis=1)
```

```python
df['year'] = Time_in_datetime.dt.year
df['month'] = Time_in_datetime.dt.month
df['day'] = Time_in_datetime.dt.day
```

```python
del Time_in_datetime
```

```python
marco_a_maio = df.loc[(df.month>=3)&\
                     (df.month<=5),:].copy()

agos_a_outubro = df.loc[(df.month>=8)&\
                     (df.month<=10), :].copy()
```
