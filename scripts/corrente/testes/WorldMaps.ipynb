### the first step is to import the basemap and matplotlib.pyplot libraries

```python
from mpl_toolkits.basemap import Basemap
from matplotlib.patches import Polygon
import matplotlib.pyplot as plt
import numpy as np
```

plotando um mapa simples com o basemap (apenas o contorno dos continentes)

```python
m = Basemap()# receives basemap function
m.drawcoastlines()# drawing the coast lines of the continents
plt.show()
```

#### Now let's colorize our map to make it a little more charming

```python
m.drawcoastlines(color='black')# ploting coast lines
m.fillcontinents(color='Darkorange'); # coloring the continents with the orange color
m.drawmapboundary(fill_color='skyblue');# coloring the ocean in blue
```

### let's see now other types of projections

Cylindrical Equidistant projection

o lon_0 e o lat_0 refer to the center of the map!

```python
m = Basemap(projection='aeqd',lon_0=-38,lat_0=-3);# defining a map with equidistant projection
m.drawmapboundary(fill_color='skyblue');#outline of the oceans with the color blue
m.fillcontinents(color='Darkorange',lake_color='skyblue');# coloring the continents orange
plt.show()
```

### Let's now plot a map with geographic coordinates

```python
#  map cylindrical projection
m = Basemap(projection='cyl',lon_0=0,
            lat_0=0,lat_ts=0,
            llcrnrlat=-90,urcrnrlat=90,
            llcrnrlon=-180,urcrnrlon=180)
m.drawcoastlines();
m.drawmapboundary(fill_color='skyblue')
m.fillcontinents(color='burlywood')# by coloring the continents
m.drawcountries()
m.drawmeridians(range(-180,180,60),
                labels=[0,0,1,]);
m.drawparallels(range(-90,90,30),
                labels=[1,0,0])
plt.show()
```
